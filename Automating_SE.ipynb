```python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("microsoft/codebert-base")
model = AutoModel.from_pretrained("microsoft/codebert-base")
```

```python
import urllib.request
import json

dataurl = "https://raw.githubusercontent.com/Meerschwein/Automating-SE/refs/heads/main/sample-1000.json"
data = json.loads(urllib.request.urlopen(dataurl).read())
```

```python
sample_function = data[0]["code"]
inputs = tokenizer(
    sample_function,
    return_tensors="pt",
    truncation=True,
    max_length=512,
    padding="max_length"
)

with torch.no_grad():
    outputs = model(**inputs)

cls_embedding = outputs.last_hidden_state[:, 0, :]

print("Raw Code:")
print(sample_function)
print("Tokenized code:")
print(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0]))
print("Function embedding shape (CLS token):")
print(cls_embedding.shape)
print("First 10 elements of embedding:")
print(cls_embedding[0][:10])
```
